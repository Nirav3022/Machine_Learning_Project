```python
weather_data = pb.read_csv("C:\Data Analyst Training\Machine_learning_Project\Wather_Prediction\seattle-weather.csv")
weather_data.head()
```

```python
weather_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

def add_label(df, column):
    df[column] = LabelEncoder().fit_transform(df[column])
    return df

weather_data = add_label(weather_data, "weather")
weather_data
        
```

```python
weather_column = ['precipitation' , 'temp_max', 'temp_min', 'wind']
```

```python
def normalize_weather(weather_data, weather_column):
    weather_data[weather_column] = weather_data[weather_column] / weather_data[weather_column].max()
    return weather_data

weather_data = normalize_weather(weather_data, weather_column)
weather_data
```

```python
drop_date = weather_data.drop("date",axis=1)
drop_date
```

```python
independent_x = weather_data.drop(["weather","date"],axis=1)
dependent_y = weather_data["weather"]
```

```python
from sklearn.model_selection import train_test_split
independent_train, independent_test, dependent_train, dependent_test = train_test_split(independent_x, dependent_y, test_size = 0.2, random_state = 0)
print("x- Tarin",independent_train)
print("y- Tarin",dependent_train)
```
